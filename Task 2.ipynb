```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
import matplotlib.pyplot as plt
```

```python
pd.crosstab(tips.sex,tips.smoker).plot(kind="bar")
```

```python
tips.day.value_counts().plot(kind='pie')
```

```python
plt.scatter(tips.smoker, tips.time)
```

```python
plt.scatter(tips.smoker, tips.day)
plt.title("smoker versus day")
plt.xlabel("smoker")
plt.ylabel("day")
```

```python
plt.hist(tips['tip'])
```

```python
plt.boxplot(tips['total_bill'])
```

```python
plt.violinplot(tips['tip'])
```

```python
sns.stripplot(x='day',y='tip',data=tips)
plt.show()
```

```python
sns.swarmplot(x='tip',y='smoker',data=tips)
plt.show()
```

```python
sns.violinplot(x='total_bill',y='smoker',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='smoker',data=tips)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
corr = tips.corr()
sns.heatmap(corr)
```
